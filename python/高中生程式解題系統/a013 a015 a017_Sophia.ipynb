```python
#013
def translate_to_numbers(num):
   
    numbers = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
   
    x1_list = []
    
    sum1 = 0
    
    for i in num:
        x1_list.append(i)
  
    for i in range(len(x1_list)):  # 0 ~ x1_list end (int)
        x1_list[i] = numbers[x1_list[i]]
   
    for i in range(1, len(x1_list)):  # 1 ~ x1_list end (int)
        if x1_list[i - 1] < x1_list[i]: # 後項大於前項
            x1_list[i - 1] = -x1_list[i - 1]
    for i in range(len(x1_list)):
        sum1 += x1_list[i]
   
    return sum1


def roma(num):
    '''將阿拉伯數字轉換成羅馬數字'''
  
    x = num

   
    thousand = x // 1000
    hundred = x % 1000 // 100
    one = x % 10
    ten = (x % 100 - one) // 10
    
    
    if thousand != 0:
        for i in range(thousand):
            print('M', end="")
  
    if hundred != 0:
        
        if hundred <= 3:
            for i in range(hundred):
                print('C', end="")
        if hundred == 4:
            print("CD", end="")
        if hundred >= 5 and hundred != 9:
            print("D", end="")
            for i in range(hundred - 5):
                print("C", end="")
        if hundred == 9:
            print("CM", end="")
   
    if ten != 0:
        # ten output
        if ten <= 3:
            for i in range(ten):
                print('X', end="")
        if ten == 4:
            print("XL", end="")
        if ten >= 5 and ten != 9:
            print("L", end="")
            for i in range(ten - 5):
                print("X", end="")
        if ten == 9:
            print("XC", end="")
   
    if one != 0:
        
        if one <= 3:
            for i in range(one):
                print('I', end="")
        if one == 4:
            print("IV", end="")
        if one >= 5 and one != 9:
            print("V", end="")
            for i in range(one - 5):
                print("I", end="")
        if one == 9:
            print("IX", end="")


while 1:

    x1 = input()
   
    if x1 == '#':
        break
    
    question = x1.split()
    x1 = question[0]
    x2 = question[1]
    
   
    ans1 = translate_to_numbers(x1) - translate_to_numbers(x2)
   
    if ans1 < 0:
        ans1 = -ans1
   
    if ans1 == 0:
        print("ZERO", end="")
    else:
        roma(ans1)
   
    print()
```

```python
#a015
import sys
mat = [[0] * 100 for i in range(100) ]  
for line in sys.stdin:
    m, n = line.split()
    m = int(m)
    n = int(n)
    for i in range(m): 
        list = sys.stdin.readline().split()
        for j in range(n):
            mat[i][j] = list[j]
    for i in range(n):  
        for j in range(m):
            print(mat[j][i]," ",sep="",end="")
        print()
```
